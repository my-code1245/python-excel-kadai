```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill # Added import

# Load 2022 data
wb_1 = openpyxl.load_workbook('2022_年間売上表.xlsx')
sheet_2022 = wb_1['Sheet1']
data_2022_generator = sheet_2022.values
headers_2022 = next(data_2022_generator) # Extract header row
df_2022 = pd.DataFrame(data_2022_generator, columns=headers_2022)

# Load 2023 data
wb_2 = openpyxl.load_workbook('2023_年間売上表.xlsx')
sheet_2023 = wb_2['Sheet1']
data_2023_generator = sheet_2023.values
headers_2023 = next(data_2023_generator) # Extract header row
df_2023 = pd.DataFrame(data_2023_generator, columns=headers_2023)

# Concatenate DataFrames
df_2022_2023 = pd.concat([df_2022, df_2023], ignore_index=True)

# Group by '売上年' and '商品', then sum '金額（千円）'
df_grouped = df_2022_2023.groupby(['売上年', '商品'], as_index=False)['金額（千円）'].sum()

df_grouped = df_grouped[['商品','売上年','金額（千円）']]

df_grouped_2 = df_grouped.sort_values(by='商品').reset_index(drop=True)

print(df_grouped_2)

# Save the DataFrame to a new Excel file
output_filename = '売上集計表.xlsx'
df_grouped_2.to_excel(output_filename, index=False)

# Load the workbook and select the active sheet
wb_new = openpyxl.load_workbook(output_filename)
ws_new = wb_new.active

# Define the gray fill
gray_fill = PatternFill(start_color='F2F2F2', end_color='F2F2F2', fill_type='solid')

# Apply the fill to the header row (first row)
for cell in ws_new[1]:
    cell.fill = gray_fill

# Save the modified workbook
wb_new.save(output_filename)
```
